```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(15,5), gridspec_kw={'width_ratios': [2,2,2,0.3],'height_ratios': [1]})

image_path = "/mnt/ncshare/ozkilim/lymphoma/CLAM/heatmaps/final_maps/heatmap_production_results_BEVNET2X_selected/HEATMAP_OUTPUT/LUAD/Her2Neg_Case_68_orig_1.jpg"
image = plt.imread(image_path)
# Create a new figure and plot the image
axes[0].imshow(image)
# Remove the axes
axes[0].axis('off')
axes[0].set_title('a. Raw Her2Neg_Case_68 WSI')



image_path = "/mnt/ncshare/ozkilim/lymphoma/CLAM/HER2_roi_vis/final_maps/Her2Neg_Case_68imagenet.png"
image = plt.imread(image_path)
# Create a new figure and plot the image
axes[1].imshow(image)
# Remove the axes
axes[1].axis('off')
axes[1].set_title('b. ImageNet SSL attentions')


# Create dummy heatmap data
heatmap_data = np.random.rand(0, 1)

# Display the heatmap using a color map (e.g., 'jet' colormap)
heatmap = axes[3].imshow(heatmap_data, cmap='jet', interpolation='nearest',alpha=0.4)
axes[3].set_visible(False)

# Add a colorbar to indicate intensity
cbar = fig.colorbar(heatmap, ax=axes[3])  # Add the colorbar
cbar.set_label('Attention weights')


image_path = "/mnt/ncshare/ozkilim/lymphoma/CLAM/HER2_roi_vis/final_maps/Her2Neg_Case_68.png"
image = plt.imread(image_path)
# Create a new figure and plot the image
axes[2].imshow(image)
# Remove the axes
axes[2].axis('off')
axes[2].set_title('c. BirdsEyeViewNet attentions')

axes[1].annotate('Pathologist annotations', xy=(1950, 1000), xytext=(200, 200),
            arrowprops=dict(facecolor='black', shrink=0.01,width=0.02, headwidth=3))

plt.suptitle("Whole slide image attention-based heatmaps",fontsize=18)

# axes[0].text(0, 0, 'a.', ha='left', va='bottom',fontsize=12, transform=axes[0].transAxes)
# axes[1].text(0, 0, 'b.', ha='left', va='bottom',fontsize=12, transform=axes[1].transAxes)
# axes[2].text(0, 0, 'c.', ha='left', va='bottom',fontsize=12, transform=axes[2].transAxes)

plt.tight_layout()

fig.savefig('../figures/heatmaps.png', dpi=500 ,bbox_inches='tight', pad_inches=.0, facecolor='white')
```
