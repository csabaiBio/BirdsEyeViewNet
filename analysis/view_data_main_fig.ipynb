### Main paper figure with datasets visualization.

```python
# Take h5 files and pull out data
fig, axes = plt.subplots(6, 2, figsize=(6, 2))
fig.tight_layout(pad=0)
plt.subplots_adjust(wspace=0.1, hspace=0)
```

```python
def plot_on_ax(h5_path,row):
    with h5py.File(h5_path, "r") as f:
        # Print all root level object names (aka keys) 
        data = np.array(f["imgs"])
    # Loop over the images and display them
    for i in range(9):
        ax = axes[row,i]  # Get the appropriate subplot
        raw_image = data[i+14,:,:,:]
        raw_image = np.transpose(raw_image)
        ax.imshow(raw_image)  # Display the image
        ax.axis('off')  # Turn off the axis labels


h5_path = "../20X_IMAGES/DLBCL/h5_files/13958_1.h5"
row = 0
plot_on_ax(h5_path,row)

h5_path = "../10X_IMAGES/Ovarian/h5_files/1612595C.h5"
row = 1
plot_on_ax(h5_path,row)

h5_path = "../20X_IMAGES/BRACS/h5_files/BRACS_283.h5"
row = 2
plot_on_ax(h5_path,row)

h5_path = "../20X_IMAGES/HunCRC/h5_files/012.h5"
row = 3
plot_on_ax(h5_path,row)

h5_path = "../20X_IMAGES/Yale/h5_files/Her2Neg_Case_05.h5"
row = 4
plot_on_ax(h5_path,row)

h5_path = "../20X_IMAGES/PANDA/h5_files/74340eaeb135c761155f1a7f198f17dc.h5"
row = 5
plot_on_ax(h5_path,row)


fig.suptitle("WSI patch samples for various cancer types.",y=1.05, fontsize=20)

axes[0,4].set_title("DLBCL: 20X", fontsize=13)

axes[1,4].set_title("Ovarian: 10X", fontsize=13)

axes[2,4].set_title("BRACS: 20X", fontsize=13)

axes[3,4].set_title("HunCRC: 20X", fontsize=13)

axes[4,4].set_title("Yale: 20X", fontsize=13)

axes[5,4].set_title("PANDA: 10X", fontsize=13)

# make rows labels..
plt.tight_layout()
```
