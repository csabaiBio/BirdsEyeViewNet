```python
import matplotlib.pylab as plt 
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from matplotlib.patches import Rectangle
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(30,10))

image_path = "/mnt/ncshare/ozkilim/lymphoma/CLAM/HER2_roi_vis/final_maps/Her2Neg_Case_36.png"
image = plt.imread(image_path)
# Create a new figure and plot the image
x0,x1,y0,y1 = 5000,6000,600,1600
axes[0].imshow(image[y0:y1,x0:x1])
rect = Rectangle((x0, y0), x1 - x0, y1 - y0, linewidth=2, edgecolor='r', facecolor='none')
axins = inset_axes(axes[0], width='40%', height='40%', loc='upper left')
axins.axis('off')
axins.add_patch(rect)
axins.imshow(image)
# Remove the axes
axes[0].axis('off')



image_path = "/mnt/ncshare/ozkilim/lymphoma/CLAM/HER2_roi_vis/final_maps/Her2Neg_Case_49.png"
image = plt.imread(image_path)
# Create a new figure and plot the image
x0,x1,y0,y1 = 800,1800,2000,3000
axes[1].imshow(image[y0:y1,x0:x1])
rect = Rectangle((x0, y0), x1 - x0, y1 - y0, linewidth=2, edgecolor='r', facecolor='none')
axins = inset_axes(axes[1], width='50%', height='50%', loc='upper left')
axins.axis('off')
axins.add_patch(rect)
axins.imshow(image)
# Remove the axes
axes[1].axis('off')


image_path = "/mnt/ncshare/ozkilim/lymphoma/CLAM/HER2_roi_vis/final_maps/Her2Neg_Case_68.png"
image = plt.imread(image_path)
# Create a new figure and plot the image
x0,x1,y0,y1 = 1500,2500,700,1700
axes[2].imshow(image[y0:y1,x0:x1])
rect = Rectangle((x0, y0), x1 - x0, y1 - y0, linewidth=2, edgecolor='r', facecolor='none')
axins = inset_axes(axes[2], width='40%', height='40%', loc='upper left')
axins.axis('off')
axins.add_patch(rect)
axins.imshow(image)
# Remove the axes
axes[2].axis('off')



plt.suptitle("Attention weights and human annotation",fontsize=14)
```
