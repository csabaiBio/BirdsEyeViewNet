```python
import pandas as pd 
import matplotlib.pylab as plt 
```

#### A script to visualise the loss curves of dino for different datasets  

```python
df = pd.read_csv("/mnt/ncshare/ozkilim/dino/saved_models/imagenet800k/log.txt",header=None,sep=" ")

df2 = pd.read_csv("/mnt/ncshare/ozkilim/dino/saved_models/satellite2M_whitepatchesdropped/log.txt",header=None,sep=" ")
```

```python
plt.figure(figsize=(8, 8))

lrs = df[1].str[:-2].astype(float)
epocs = df[7].str[:-1].astype(float)
plt.scatter(epocs,lrs, label="Imagenet 800k")
lrs = df2[1].str[:-2].astype(float)
epocs = df2[7].str[:-1].astype(float)
plt.scatter(epocs,lrs, label="satelite 2M")
plt.title("Dino loss")
plt.legend()
plt.ylabel("Loss")
plt.xlabel("Epocs")
```
